## Cleaning the Data

```python
#Finding the sum of null values for each attribute
pd.isnull(rd).sum()
```

## Dropping duplicate rows

```python
rd.drop_duplicates(["movie_title"],inplace=True, keep='first')
```

```python
print(rd.duplicated("movie_title").sum())
```

### Removing unwanted Features

```python
#features which won't affect our prediction of box office revenue
unwanted=["movie_imdb_link","content_rating","facenumber_in_poster","aspect_ratio","Color","director_facebook_likes","actor_3_facebook_likes","plot_keywords","actor_2_facebook_likes","actor_1_facebook_likes","cast_total_facebook_likes","movie_facebook_likes"]
```

```python
#getting the rows that have at least 13 features as null
rd = rd[rd.isnull().sum(axis=1)<=13]
```

```python
rd = rd.reset_index()
rd
```

```python
rd.drop(unwanted,inplace=True,axis=1)
```

```python
rd.columns
```

## Dropping ROWS with missing values for the Features which cannot be predicted 

```python
#using dropna, passing a set of all such columns
rd = rd.dropna(subset=["director_name","country","actor_2_name","actor_1_name","actor_3_name","title_year"])
```

## Using mean and mode to get a prediction of the values of the missing values

```python
#below given feutures with missing values can be predicted
rd["num_critic_for_reviews"].fillna(rd["num_critic_for_reviews"].mean(),inplace=True)
rd["num_user_for_reviews"].fillna(rd["num_user_for_reviews"].mean(),inplace=True)
rd["duration"].fillna(rd["duration"].mean(),inplace=True)
rd["budget"].fillna(rd["budget"].mean(),inplace=True)
rd["gross"].fillna(rd["gross"].mean(),inplace=True)

rd.reset_index(inplace=True,drop=True)
```

### Turing all character data to upper case


```python
rd["director_name"]=rd["director_name"].str.upper()
rd["actor_2_name"] = rd["actor_2_name"].str.upper()
rd["actor_1_name"] = rd["actor_1_name"].str.upper()
rd["actor_3_name"] = rd["actor_3_name"].str.upper()
rd["language"] = rd["language"].str.upper()
rd["country"]=rd["country"].str.upper()
rd["genres"]=rd["genres"].str.upper()
rd["movie_title"]=rd["movie_title"].str.upper()
rd["language"]= rd["language"].str.upper()

language_missing = rd[["country","language"]].groupby(["country"]).agg(pd.Series.mode).reset_index()
country_language_d = dict(zip(language_missing["country"],language_missing["language"]))
rd["language"].fillna(rd["country"].map(country_language_d),inplace=True)
```

```python
genre_count = dict()
x = rd["genres"].str.split("|")
for y in x:
     for z in y:
        if z not in genre_count:
            genre_count[z]=1
        else:
            genre_count[z] += 1

genre_data = pd.DataFrame(genre_count.items(),columns=["Genre","Count"])
```

```python
genre_data=genre_data.sort_values("Count",ascending=False).head(10).reset_index()
genre_data.drop("index",axis=1,inplace=True)
genre_data
```

```python
for x in genre_data["Genre"]:
    y = rd.loc[rd["genres"].str.find(x)!=-1]["imdb_score"].idxmax()
    z = rd.loc[rd["genres"].str.find(x)!=-1]["gross"].idxmax()
    genre_data.loc[genre_data["Genre"]==x,"Best Movie_imdb"]=rd.loc[y,"movie_title"]
    genre_data.loc[genre_data["Genre"]==x,"Max IMDB Score"]=rd.loc[y,"imdb_score"]
    genre_data.loc[genre_data["Genre"]==x,"Best Movie_gross"]=rd.loc[z,"movie_title"]
    genre_data.loc[genre_data["Genre"]==x,"Gross"]=rd.loc[z,"gross"]

genre_data
```

## Duration of Movies

```python
shorts=0;
feat=0;
reg=0;
long=0;
total=0;
for i in rd.duration:
    if (i<=60):
        shorts+=1
    elif (i>60 and i<=90):
        feat+=1
    elif (i>90 and i<=180):
        reg+=1
    else:
        long+=1
    total+=1
dur=[shorts,feat,reg,long]
```

```python
plt.style.use("ggplot")
labels=["short(under 60 mins)","Feature(60-120 mins)","Regular(120-180 mins)","Long(above 180 mins)"]
explode=(0.4,0.1,0,0.5)
plt.figure(figsize=(10,20), dpi=80)
plt.pie(dur, labels=labels, autopct='%.2f %%', explode=explode,pctdistance=2)


plt.title("Movie Duration Percentage", fontdict={'fontsize':50, 'fontweight':'bold'})

plt.savefig('Movie Duration Percentage.png')
plt.show()
```
